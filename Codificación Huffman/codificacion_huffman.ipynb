```
import numpy as np

probabilidad_letra = {'a': None, 'b': None, 'c': None, 'd': None, 'e': None, 'f': None, 'g': None,
                      'h': None, 'i': None, 'j': None, 'k': None, 'l': None, 'm': None, 'n': None,
                      'ñ': None, 'o': None, 'p': None, 'q': None, 'r': None, 's': None, 't': None,
                      'u': None, 'v': None, 'w': None, 'x': None, 'y': None, 'z': None}

print('Cantidad de letras:', len(probabilidad_letra.keys()))

# Generar una lista de 27 números aleatorios entre 0 y 1
random_numbers = np.random.rand(27)

# Normalizar la lista para que la suma sea igual a 1
normalized_numbers = random_numbers / sum(random_numbers)

list_claves = list(probabilidad_letra.keys())

# Agregar las probabilidades al diccionario
for i in range(len(list_claves)):
  probabilidad_letra[list_claves[i]] = normalized_numbers[i]

for c, v in probabilidad_letra.items():
  print(c, v)

print('\nSuma de las probabilidades:', sum(probabilidad_letra.values()))
```

```
# Ordenar las letras en funcíon de la probabilidad asignada
import pandas as pd

pd.set_option('display.max_columns', None)

df = pd.DataFrame(probabilidad_letra, index=['P0'])
df_transpuesta = df.transpose()
df_sorted = df_transpuesta.sort_values(by='P0', ascending=False)
df_sorted
```
